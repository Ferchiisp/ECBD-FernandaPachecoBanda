```python
#Limpieza de datos
# Convertir nombres de columnas a minúsculas
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
# Identificar valores nulos
df.isnull().sum()
```

```python
# Calcular porcentaje de valores faltantes
(df.isnull().sum() / len(df)) * 100
```

```python
# Detectar registros duplicados
df.duplicated().sum()
```

```python
# Eliminar registros duplicados
df.drop_duplicates(inplace=True)
```

```python
# Rellenar valores nulos en director
df["director"] = df["director"].fillna("Desconocido")
```

```python
# Rellenar valores nulos en cast
df["cast"] = df["cast"].fillna("No registrado")
```

```python
# Rellenar valores nulos en country
df["country"] = df["country"].fillna("No especificado")
```

```python
# Rellenar valores nulos en rating con la moda
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
# Rellenar valores nulos en duration
df["duration"] = df["duration"].fillna("Sin duración")
```

```python
# Verificar nuevamente los valores nulos
df.isnull().sum()
```

```python
# JUSTIFICACIÓN DE LAS DECISIONES
# 1. Se eliminaron registros duplicados para evitar datos repetidos.
# 2. Se rellenaron valores faltantes con texto para conservar la mayor cantidad de información posible.
# 3. En rating se utilizó la moda porque representa el valor más frecuente.
# 4. Los nombres de columnas se estandarizaron en minúsculas para facilitar el análisis y evitar errores.
```
